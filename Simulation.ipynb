- 1 define the function
- 2 set the condition
- 3 use the 45 line to generate the list of numbers
- 4 plot the result while iteration

```python
import matplotlib.pyplot as plt
import math
```

```python
#a = 3.414
def fo(a, x_current): # the first order difference equation
    y_current = x_current * a * (1-x_current)
   #y_current = x_current * math.exp(a*(1-x_current))
    return y_current

def l_45(y_current): # the relationship between current y and next x
    x_next = y_current
    return x_next
```

```python
def simulation_visual(a_value, x_min, x_max,iteration_times):
    a = a_value
    x_current = x_min
    plt.scatter(x_current, fo(a, x_current), s=1, c="blue")
    
    for i in range(iteration_times):
        y_current = fo(a, x_current)
        # print(x_current, y_current)
        # Skip the fixed point where fo(x_current) == l_45(x_current)
        if abs(x_current - y_current)<1e-6:
            x_current += 0.01  # Avoid being stuck in a fixed point
        else:
            x_current = l_45(y_current)
        # Plot current state
        plt.scatter(x_current, l_45(x_current), s=10, c='red')
        plt.scatter(x_current, fo(a, x_current), s=1, c="blue")

    plt.xlabel('x')
    plt.ylabel('fo(x)')
```

```python
simulation_visual(2.707, 0, 1, 200)
plt.show()
# simulation_visual(0.9, 0, 5)
```

```python
simulation_visual(3.83, 0, 1 ,800)
```

simulate the F(F(x))

```python
def fo(a, x_current): # the first order difference equation
    y_current = x_current * a * (1-x_current)
    return y_current
def fo_2(a, x_current):
    fo_result = fo(a,x_current)
    fo_2_result = fo(a,fo_result)
    return fo_2_result
def l_45(y_current): # the relationship between current y and next x
    x_next = y_current
    return x_next
```

```python
def simulation_visual_2(a_value, x_min, x_max):
    a = a_value
    x_current = x_min
    plt.scatter(x_current, fo_2(a, x_current), s=1, c="blue")
    
    for i in range(1000):
        y_current = fo_2(a, x_current)
        # Skip the fixed point where fo(x_current) == l_45(x_current)
        if x_current == y_current:
            x_current += 0.01  # Avoid being stuck in a fixed point
        else:
            x_current = l_45(y_current)
        # Plot current state
        plt.scatter(x_current, l_45(x_current), s=10, c='red')
        plt.scatter(x_current, fo_2(a, x_current), s=1, c="blue")

    plt.xlabel('x')
    plt.ylabel('fo(x)')
```

```python
simulation_visual_2(3.6, 0, 1)
plt.show()
```
